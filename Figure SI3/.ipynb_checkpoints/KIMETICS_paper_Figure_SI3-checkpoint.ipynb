```python
import pandas as pd
import pathlib
import h5py
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import curve_fit
import numpy as np
import tkinter as tk
from tkinter import filedialog
from scipy.ndimage import gaussian_filter1d
from scipy.optimize import curve_fit
from scipy.signal import find_peaks


def gaussian(x, A, mu, sigma):
    return A * np.exp(-(x - mu)**2 / (2 * sigma**2))
    
def two_gaussians(x, A1, mu1, sigma, A2, mu2):
    return (gaussian(x, A1, mu1, sigma) + gaussian(x, A2, mu2, sigma))

def select_file():
    root = tk.Tk()
    root.withdraw()

    file_path = filedialog.askopenfilename()

    if file_path:
        return pathlib.Path(file_path)
    else:
        return None

systems=[r'NspeNspeNazo + H$^+$', r'NspeNspeNazo + Na$^+$', r'NspeNspeN2tz + H$^+$', r'NspeNspeN2tzNspe + H$^+$']
data=[]
paths=[Path_to_files]
for i, path in enumerate(paths):
    DIR=pathlib.Path(path)
    df = pd.read_csv(DIR, delimiter='\t')
    df = df.drop(index=0)
    df = df.reset_index(drop=True)
    df = df.astype(float)
    df['INorm(full)']=df['INorm(full)']-df['INorm(full)'].min()
    df['INorm(full)_normalized'] = (df['INorm(full)'] - df['INorm(full)'].min()) / (df['INorm(full)'].max() - df['INorm(full)'].min())
    if i==2 :
        df = df[(df['Arrival Time'] >= 14) & (df['Arrival Time'] <= 17.5)]
    else:
        df = df[(df['Arrival Time'] >= 12) & (df['Arrival Time'] <= 21)]
    x, y=df['Arrival Time'], df['INorm(full)']
    data.append((x, y))

fig, axes = plt.subplots(2, 2, figsize=(7, 5), dpi=600)
sns.set(style="whitegrid")

for i, (x, y) in enumerate(data):
    ax = axes.flat[i]
    
    x=np.array(x)
    y=np.array(y)
    
    peaks, _ = find_peaks(y, height=0)
    peak_x=x[peaks]
    peak_y=y[peaks]
    if len(peak_x) >= 2:
        sorted_indices = np.argsort(peak_y)[-2:]
        max_x = peak_x[sorted_indices]
    bounds=[[0, -np.inf, -np.inf, 0, -np.inf], [np.inf, np.inf, np.inf, np.inf, np.inf]]
    guess=[np.max(y), max_x[0], 0.2, np.max(y),max_x[1]]
    popt, _ = curve_fit(two_gaussians, x, y, p0=guess, bounds=bounds)
    if i==0:
        Fig=popt
    ax.scatter(x, y, color='black', s=10, label='Data', alpha=0.6)
    x=np.linspace(min(x), max(x), 2000)
    y_fit = two_gaussians(x, *popt)
    y_gaussian1 = gaussian(x, popt[0], popt[1], popt[2])
    y_gaussian2 = gaussian(x, popt[3], popt[4], popt[2])
    ax.plot(x, y_gaussian1, 'g--', label='Gaussian 1')
    ax.plot(x, y_gaussian2, 'g--', label='Gaussian 2')
    ax.plot(x, y_fit, 'r-', label='Fit (Two Gaussians)', linewidth=2)
    ax.set_title(f'{systems[i]} system', fontsize=10)
    ax.set_xlabel('Arrival time (ms)', fontsize=10)
    ax.set_ylabel('Intensity (arb. units)', fontsize=10)
    ax.set_facecolor('#F8F8FB')
    ax.grid(False)


plt.tight_layout(pad=2.0)
plt.show()
```
