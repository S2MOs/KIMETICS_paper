```python
def gamma_Vso(x):
    return 7E-5*x**2

fig, ax=plt.subplots(figsize=[7, 5], dpi=600)
sns.set(style="whitegrid")
x=np.linspace(0, 110, 2000)
y=gamma_Vso(x)
sns.lineplot(x=x, y=y)
ax.set_facecolor('#F8F8FB')
ax.grid(False)
ax.set_xlabel('Spin-orbit coupling V$_{SO}$ (cm$^{-1}$)')
ax.set_ylabel('Eyring transmission coefficient')
plt.savefig(pathlib.Path(r'Gamma_Vso_Figure.png'), dpi=600)
plt.savefig(pathlib.Path(r'Gamma_Vso_Figure.svg'), dpi=600)
plt.show()
```
