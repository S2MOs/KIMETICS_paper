```python
from KIMETICS import *
from matplotlib.ticker import ScalarFormatter
```

```python
DIR=pathlib.Path(r'Path_to_files')
df1=pd.read_csv(DIR/pathlib.Path("SYNAPT_THRO_241115_LCMS_NspeNspeN2tz_NH4I_20deg_1_kinetic.csv"))
df2=pd.read_csv(DIR/pathlib.Path("SYNAPT_THRO_241115_LCMS_NspeNspeN2tz_NH4I_20deg_2_kinetic.csv"))
fig, ax=plt.subplots(figsize=[7, 5], dpi=600)
sns.set(style="whitegrid")
sns.scatterplot(data=df1, x='Time', y='Z', label='0 eq. NH$_{4}$I', alpha=0.6)
x = df1['Time'].to_numpy()
y = df1['Z'].to_numpy()
popt, perr, rep, r2 = fit_exponential_decay(x, y, y0=0)
print(rep)
x_fit = np.linspace(x.min(), x.max(), 1000)
y_fit = exponential_decay(x_fit, *popt)
sns.lineplot(x=x_fit, y=y_fit, label=f'R$^{2}$ = {r2.round(3)}')
sns.scatterplot(data=df2, x='Time', y='Z', label='100 eq. NH$_{4}$I', alpha=0.6)
x = df2['Time'].to_numpy()
y = df2['Z'].to_numpy()
popt, perr, rep, r2 = fit_exponential_decay(x, y, y0=0)
print(rep)
x_fit = np.linspace(x.min(), x.max(), 1000)
y_fit = exponential_decay(x_fit, *popt)
sns.lineplot(x=x_fit, y=y_fit, label=f'R$^{2}$ = {r2.round(3)}')
ax.set(xlabel="Time (s)", ylabel="Z-isomer abundance")
ax.set_facecolor('#F8F8FB')
ax.xaxis.set_major_formatter(ScalarFormatter(useMathText=True))
ax.ticklabel_format(style='sci', axis='x', scilimits=(0, 0))
plt.tight_layout()
plt.grid(False)
plt.show()
```
