```python
import pandas as pd
import pathlib
import h5py
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import curve_fit
import numpy as np
import tkinter as tk
from tkinter import filedialog
from scipy.ndimage import gaussian_filter1d
from scipy.optimize import curve_fit
from scipy.signal import find_peaks

def extract_RT(df) :
    df_copy = df.copy()
    df_copy = df_copy.groupby("retention_time").sum()
    extracted_df = pd.DataFrame()
    extracted_df["retention_time"] = np.array(df_copy.index)
    extracted_df["intensity"] = df_copy["intensity"].values
    return extracted_df

def filter_MZ(df, MZ, dMZ) :
    df_copy = df.copy()
    lower_bound = MZ - dMZ
    upper_bound = MZ + dMZ
    df_copy.loc[(df_copy['mz'] < lower_bound) | (df_copy['mz'] > upper_bound), 'intensity'] = 0
    return df_copy
    
def gaussian(x, A, mu, sigma):
    return A * np.exp(-(x - mu)**2 / (2 * sigma**2))

def two_gaussians(x, A1, mu1, sigma1, A2, mu2, sigma2):
    return (gaussian(x, A1, mu1, sigma1) + gaussian(x, A2, mu2, sigma2))

def select_file():
    root = tk.Tk()
    file_path = filedialog.askopenfilename()
    root.withdraw()
    if file_path:
        return pathlib.Path(file_path)
    else:
        return None

systems=[r'NspeNspeNazo', r'NspeNspeN2tz', r'NspeNspeN2tzNspe']
MZ=[577.3, 584.3, 745.3]
data=[]
paths=[Path_to_files]

fig, axes = plt.subplots(2, 2, figsize=(7, 5), dpi=600)
sns.set(style="whitegrid")
fig.delaxes(axes[1, 1])

for i, path in enumerate(paths):
    ax = axes.flat[i]
    
    df=pd.read_hdf(path)
    df=filter_MZ(df, MZ[i], 1)
    df=extract_RT(df)
    df = df[(df['retention_time'] >= 2) & (df['retention_time'] <= 7)]
    df['intensity'] = (df['intensity'] - df['intensity'].min()) / (df['intensity'].max() - df['intensity'].min())
    x=np.array(df['retention_time'])
    y=np.array(df['intensity'])
    
    sigma_smoothing = 1.0  
    y = gaussian_filter1d(y, sigma=sigma_smoothing)
    
    peaks, _ = find_peaks(y, height=0)
    peak_x=x[peaks]
    peak_y=y[peaks]
    if len(peak_x) >= 2:
        sorted_indices = np.argsort(peak_y)[-2:]
        max_x = peak_x[sorted_indices]
    sigma_smoothing = 1.0  
    y = gaussian_filter1d(y, sigma=sigma_smoothing)

    bounds=[[0, -np.inf, -np.inf, 0, -np.inf, -np.inf], [np.inf, np.inf, np.inf, np.inf, np.inf, np.inf]]
    guess=[np.max(y), max_x[0], 0.2, np.max(y),max_x[1], 0.2]
    popt, _ = curve_fit(two_gaussians, x, y, p0=guess, bounds=bounds)
    
    if popt[1]>popt[4]:
        x_min=popt[4]-1
        x_max=popt[1]+1
    else:
        x_min = popt[1]-1
        x_max = popt[4]+1

    mask = (x >= x_min) & (x <= x_max)
    x = x[mask]
    y = y[mask]
    ax.scatter(x, y, color='black', s=10, label='Data', alpha=0.6)
    x=np.linspace(min(x), max(x), 2000)
    y_fit = two_gaussians(x, *popt)
    y_gaussian1 = gaussian(x, popt[0], popt[1], popt[2])
    y_gaussian2 = gaussian(x, popt[3], popt[4], popt[5])
    ax.plot(x, y_gaussian1, 'g--', label='Gaussian 1')
    ax.plot(x, y_gaussian2, 'g--', label='Gaussian 2')
    ax.plot(x, y_fit, 'r-', label='Fit (Two Gaussians)', linewidth=2)
    ax.set_title(f'Chromatogram of {systems[i]} system', fontsize=8)
    ax.set_xlabel('Retention time (min)', fontsize=8)
    ax.set_ylabel('Intensity (arb. units)', fontsize=8)
    ax.set_facecolor('#F8F8FB')
    ax.grid(False)
    if popt[1]<popt[4]:
        ax.text(popt[1]-0.5, two_gaussians(popt[1],*popt)-0.05, f'{popt[1].round(2)} min', fontsize=7)
        ax.text(popt[4]+0.1, two_gaussians(popt[4],*popt)-0.05, f'{popt[4].round(2)} min', fontsize=7)
    else:
        ax.text(popt[4]-0.5, two_gaussians(popt[4],*popt)-0.05, f'{popt[4].round(2)} min', fontsize=7)
        ax.text(popt[1]+0.1, two_gaussians(popt[1],*popt)-0.05, f'{popt[1].round(2)} min', fontsize=7)
        
plt.tight_layout(pad=2.0)

plt.savefig(pathlib.Path(r'C:\Users\Thomas\Documents\#UMONS\FRIA\Data\Paper 1\AZO_3_LC_Figure\AZO_3_LC_Figure.png'), dpi=600)
plt.savefig(pathlib.Path(r'C:\Users\Thomas\Documents\#UMONS\FRIA\Data\Paper 1\AZO_3_LC_Figure\AZO_3_LC_Figure.svg'), dpi=600)

plt.show()
```
